```python
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt

def check_price(price):
    print('Checking the statistical distribution of prices')
    print(entire_home_avg_price['price'].describe())
    
    print('Fitting the prices into normal distribution')
    sns.distplot(price, fit=norm)

    # Get the fitted parameters used by the function
    (mu, sigma) = norm.fit(price)
    print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

    #Now plot the distribution
    plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
                loc='best')
    plt.ylabel('Frequency')
    plt.title('Price distribution')

    #Get also the QQ-plot
    fig = plt.figure()
    res = stats.probplot(price, plot=plt)
    plt.show()

def correct_price(price):
    price = np.log1p(price)

    #Check the new distribution 
    sns.distplot(price , fit=norm);

    # Get the fitted parameters used by the function
    (mu, sigma) = norm.fit(price)
    print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

    #Now plot the distribution
    plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
                loc='best')
    plt.ylabel('Frequency')
    plt.title('Price distribution')

    #Get also the QQ-plot
    fig = plt.figure()
    res = stats.probplot(price, plot=plt)
    plt.show()
    return price
```

```python
# add markers to map
def add_markers(data, myColor):
    for neighborhood, lat, lng, price in zip(data['neighbourhood'], data['latitude'], data['longitude'], data['price']):
        label = '{} Average price: {}$/night'.format(neighborhood, round(price, 2))
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            popup=label,
            color=myColor,
            fill=True,
            fill_color=myColor,
            fill_opacity=0.7,
            parse_html=False).add_to(map_newyork)  
add_markers(entire_home_avg_price, 'blue')  
add_markers(private_room_avg_price, 'red') 
add_markers(shared_room_avg_price, 'green')  
map_newyork
```

```python
def getNearbyVenues(data, radius=500, limit=100):
    
    venues_list=[]
    print('Obtaining venues around the neighbourhoods: ', end='')
    for name, lat, lng in zip(data['neighbourhood'], data['latitude'], data['longitude']):
        print('.', end='')
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            limit)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['neighbourhood', 
                  'latitude', 
                  'longitude', 
                  'Venue', 
                  'Venue Category']
    return nearby_venues
```

### Analyze each neighbourhood

```python
def one_hot_encode(venues):
    # one hot encoding
    onehot = pd.get_dummies(venues[['Venue Category']], prefix="", prefix_sep="")

    # add neighbourhood column back to dataframe
    onehot['neighbourhood'] = venues['neighbourhood'] 

    # move neighbourhood column to the first column
    fixed_columns = [onehot.columns[-1]] + list(onehot.columns[:-1])
    onehot = onehot[fixed_columns]

    # group the data by neighbourhood
    grouped = onehot.groupby('neighbourhood').mean().reset_index()
    return grouped
```

```python
entire_home_grouped = one_hot_encode(entire_home_venues)
print(entire_home_grouped.shape)
entire_home_grouped.head()
```

```python
private_room_grouped = one_hot_encode(private_room_venues)
print(private_room_grouped.shape)
private_room_grouped.head()
```

```python
shared_room_grouped = one_hot_encode(shared_room_venues)
print(shared_room_grouped.shape)
shared_room_grouped.head()
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row
    row_categories_sorted = row_categories.sort_values(ascending=False)
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10
indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['cluster_label', 'price']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

def sort_venues(clusters):
    # create a new dataframe
    venues_sorted = pd.DataFrame(columns=columns)
    venues_sorted['cluster_label'] = clusters['cluster_label']
    venues_sorted['price'] = clusters['price']

    for ind in np.arange(clusters.shape[0]):
        venues_sorted.iloc[ind, 2:] = return_most_common_venues(clusters.iloc[ind, 2:], num_top_venues)
    return venues_sorted
```
